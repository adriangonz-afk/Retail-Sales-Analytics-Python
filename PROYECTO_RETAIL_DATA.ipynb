# **LIMPIAMOS DATOS**

```python
#Rellenamos los markdows de Features
for col in ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']:
    feature[col].fillna(0, inplace=True)
#Rellenamos con la media los datos faltantes en las columnas CPI y Unemployment
feature['CPI'].fillna(feature['CPI'].mean(), inplace=True)
feature['Unemployment'].fillna(feature['Unemployment'].mean(), inplace=True)
feature.isnull().sum()
```

# **TRANSFORMACION DE DATOS**

```python
#Convertimos los datos de tipo fecha a datetime
feature['Date']=pd.to_datetime(feature['Date'], format='%d/%m/%Y')
feature.info()
```

```python
sales['Date']=pd.to_datetime(sales['Date'], format='%d/%m/%Y')
sales.info()
```

```python
#Creamos la columnda TotalMarkdown
def sumaMarkdowns(row):
    # row es una pd.Series con ndice MarkDown1MarkDown5
    return row.sum()

feature['TotalMarkdown'] = feature[
    ['MarkDown1','MarkDown2','MarkDown3','MarkDown4','MarkDown5']].apply(sumaMarkdowns, axis=1)

#Creamos nuevas columnas apartir del Date
# Con el metodo 'dt' puedo extraer diferentes datos de esa fecha como el ao, mes ,
#dia , dia de semana, hora, minuto ,segundo ,etc.
feature['Year'] = feature['Date'].dt.year
feature['Month'] = feature['Date'].dt.month
feature['Day'] = feature['Date'].dt.day
feature['Day_Name']= feature['Date'].dt.day_name()
feature.head(10)
#Estos datos extraidos nos sirven luego para filtrar o agrupar datos para saber informacion relevante
```

```python
#Creamos nuevas columnas apartir del Date
# Con el metodo 'dt' puedo extraer diferentes datos de esa fecha como el ao, mes ,
#dia , dia de semana, hora, minuto ,segundo ,etc.
sales['Year'] = sales['Date'].dt.year
sales['Month'] = sales['Date'].dt.month
sales['Day'] = sales['Date'].dt.day
sales['Day_Name']= sales['Date'].dt.day_name()
sales.head(10)
#Estos datos extraidos nos sirven luego para filtrar o agrupar datos para saber informacion relevante
```

```python
stores_sales = pd.merge(sales,stores, on='Store', how='inner')
sales_by_store_type =stores_sales.groupby('Type')['Weekly_Sales'].sum().reset_index(name='Weekly_Sales')
sales_by_store_type
```

```python
sales_by_store = stores_sales.groupby('Store')['Weekly_Sales'].sum().reset_index(name='TotalSales')
sales_by_store= sales_by_store.merge(stores[['Store','Type']], on= 'Store', how='left')
sales_by_store.sort_values(by='TotalSales', ascending=False, inplace=True)
sales_by_store.head(10)
```

 **PREGUNTA 2**

```python
sales_by_dept = stores_sales.groupby('Dept')['Weekly_Sales'].sum().reset_index(name='TotalSales')
sales_by_dept.sort_values(by='TotalSales', ascending=False, inplace=True)
total= sales_by_dept['TotalSales'].sum()
sales_by_dept['CumSum']   = sales_by_dept['TotalSales'].cumsum()#Acumulo la suma de las ventas de los departamentos, incrementa mientras bajas el row
sales_by_dept['CumPct']   = sales_by_dept['CumSum'] / total#Porcentaje acumulado de las ventas acumuladas

# 4) Filtrar departamentos que juntos aportan hasta el 80%
pareto_80 = sales_by_dept[sales_by_dept['CumPct'] <= 0.80].copy()
pareto_20 = sales_by_dept[sales_by_dept['CumPct'] > 0.80].copy()
print("Numero de departamentos que representan el 80% de ventas:")
pareto_80.shape[0]
```

**PREGUNTA 4**

```python
weekly = sales.groupby('Date')['Weekly_Sales'].sum().reset_index()

# C) Une el flag de feriado: busca en feature (o sales si ya lo tiene)
#    Aqu suponemos que sales ya incluye la columna IsHoliday
weekly = weekly.merge(
    sales[['Date','IsHoliday']].drop_duplicates(),
    on='Date',
    how='left'
)
```

```python
# A) Calcula un promedio mvil de 4 semanas
weekly['MA4'] = weekly['Weekly_Sales'].rolling(window=4, center=True).mean()

# B) Separa en dos DataFrames: feriados vs. no feriados
hol = weekly[weekly['IsHoliday'] == True]
nohol = weekly[weekly['IsHoliday'] == False]
```

```python
weekly['STD4'] = weekly['Weekly_Sales'].rolling(window=4, center=True).std()
weekly['Upper'] = weekly['MA4'] + weekly['STD4']
weekly['Lower'] = weekly['MA4'] - weekly['STD4']
```

```python
import matplotlib.pyplot as plt

# Prepara el eje y los datos
x = sales_by_dept['Dept'].astype(str)
y = sales_by_dept['TotalSales']
y_pct = sales_by_dept['CumPct']
idx_80 = next(i for i, v in enumerate(sales_by_dept['CumPct']) if v >= 0.8)

fig, ax1 = plt.subplots(figsize=(12,6))

# 1) Barras de TotalSales
ax1.bar(x, y)
ax1.set_ylabel('Ventas Totales en cientos de millones', fontsize=12)
ax1.set_xlabel('Departamento (Dept)', fontsize=12)
ax1.set_xticklabels(x, rotation=90)

# 2) Crear segundo eje para el porcentaje acumulado
ax2 = ax1.twinx()
ax2.plot(x, y_pct, color='red', marker='o', linewidth=2)
ax2.set_ylabel('Porcentaje Acumulado', fontsize=12)
ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda v, _: f'{v:.0%}'))

# 3) Lnea de corte al 80%
ax2.axhline(0.80, color='gray', linestyle='--', linewidth=1.5)
ax1.axvline(idx_80, color='blue', linestyle='--', linewidth=1.5, label=f'Lmite 80% en dept {x[idx_80]}')

# 4) Ttulos y layout
plt.title('Grfico de Pareto: Departamentos que concentran el 80% de Ventas', fontsize=14)
fig.tight_layout()
plt.show()
```
